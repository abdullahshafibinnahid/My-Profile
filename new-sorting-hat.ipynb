Welcome to Hogwarts. You, the first year students, will be divided into 4 houses. This is done by the old sorting hat of Hogwarts. But this year the old sorting hat has broken. As a result, a new sorting hat is needed. We will use machine learning to make it. Will you help us?

```python
import pandas as pd

from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC

train_df = pd.read_csv("/kaggle/input/datasets/maricinnamon/harry-potter-movies-dataset/Harry_Potter_Movies/Characters.csv", encoding="latin1")
train_df = train_df.dropna(subset=["House"])

test_df = pd.DataFrame([{
    "Character ID": 167,
    "Character Name": "Shafi",
    "Species": "Human",
    "Gender": "Male",
    "Patronus": "Phoenix",
    "Wand (Wood)": "Walnut",
    "Wand (Core)": "Phoenix Feather"
}])

train_df["text"] = (
    train_df["Character ID"].astype(str) + " " +
    train_df["Character Name"].fillna("") + " " +
    train_df["Species"].fillna("") + " " +
    train_df["Gender"].fillna("") + " " +
    train_df["Patronus"].fillna("") + " " +
    train_df["Wand (Wood)"].fillna("") + " " +
    train_df["Wand (Core)"].fillna("")
).str.lower().str.strip()

test_df["text"] = (
    test_df["Character ID"].astype(str) + " " +
    test_df["Character Name"].fillna("") + " " +
    test_df["Species"].fillna("") + " " +
    test_df["Gender"].fillna("") + " " +
    test_df["Patronus"].fillna("") + " " +
    test_df["Wand (Wood)"].fillna("") + " " +
    test_df["Wand (Core)"].fillna("")
).str.lower().str.strip()

X = train_df["text"]
y = train_df["House"]

model = Pipeline([

    ('tfidf', TfidfVectorizer(
        max_features=20000,
        ngram_range=(1, 2),
        stop_words='english'
    )),

    ('clf', LinearSVC())
])

model.fit(X, y)

test_pred = model.predict(test_df["text"])

submission = pd.DataFrame({
    "Character ID": test_df["Character ID"],
    "House": test_pred
})

submission.to_csv("your_house.csv", index=False)

submission.head(1)
```
